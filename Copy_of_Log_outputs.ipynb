```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np


csv_file = "/content/Logs.CSV"
entries_per_sheet = 65537  # Number of rows per sheet (fixed)
chunk_size = 3276  # Number of rows to process per frame
max_x_labels = 50  # Maximum number of x-axis labels
output_dir = "output_frames_006"
os.makedirs(output_dir, exist_ok=True)


data = pd.read_csv(csv_file, header=None)

#number of sheets
num_sheets = (len(data) + entries_per_sheet - 1) // entries_per_sheet

#cumulative data for each sheet
cumulative_data = [{"x": [], "y": []} for _ in range(num_sheets)]

# Fixed axis
fixed_x_min ="02:26:29"
fixed_x_max = "03:44:50"
fixed_y_min = 0
fixed_y_max = 100

# color cycle
colors = plt.cm.get_cmap('tab10', num_sheets)

#  sheet based on the fixed number of entries per sheet
for sheet_index in range(num_sheets):
    start_row = sheet_index * entries_per_sheet
    end_row = min(start_row + entries_per_sheet, len(data))
    sheet_data = data.iloc[start_row:end_row].sort_values(by=0)

    # Progressive plotting in chunks
    for i in range(0, len(sheet_data), chunk_size):
        chunk = sheet_data.iloc[i:i + chunk_size]
        x = chunk.iloc[:, 0].tolist()
        y = chunk.iloc[:, 5].tolist()

        #   figure for each chunk,
        fig = plt.figure(num=sheet_index + 1, clear=True)

        #  cumulative data for this sheet
        cumulative_data[sheet_index]["x"].extend(x)
        cumulative_data[sheet_index]["y"].extend(y)

        # Plotting
        if cumulative_data[sheet_index]["x"]:
            plt.plot(
                cumulative_data[sheet_index]["x"],
                cumulative_data[sheet_index]["y"],
                label=f'Sheet {sheet_index + 1}',
                color=colors(sheet_index)  #  color based
            )

        # fixed axis
        plt.xlim(fixed_x_min, fixed_x_max)
        plt.ylim(fixed_y_min, fixed_y_max)

        # Limit the number of x-axis labels to a maximum of max_x_labels
        if cumulative_data[sheet_index]["x"]:
            # Calculate step to limit the number of labels
            step = max(1, len(cumulative_data[sheet_index]["x"]) // 10)
            x_ticks = cumulative_data[sheet_index]["x"][::step]
            plt.xticks(ticks=x_ticks, labels=x_ticks, rotation=45)

        # Set the title and labels
        plt.title(f'Progressive Line Chart for Sheet {sheet_index + 1}')
        plt.xlabel('Time (Column 0)')
        plt.ylabel('Data (Column 5)')
        plt.legend(loc='upper left')

        # Save the current state of the graph as a PNG
        plt.savefig(os.path.join(output_dir, f'frame_{sheet_index:02d}_{i // chunk_size:04d}.png'))

        print(f"Saved frame for Sheet {sheet_index + 1}, Chunk {i // chunk_size}")

# print out
print(f"Saved all frames to {output_dir}")


plt.close('all')
```
